```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
from scipy.integrate import ode
import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = 16, 12  # that's default image size for this interactive session
```

```python
#Param
g = 1.
#Fixpoints
x_fix = [[0.,0.],[-1.,0.],[1.,0.]]
#Integration params

#Grid
bound = 10.
rng = np.arange(-bound,bound,0.1)
grng = np.arange(0,bound,0.05)
X,Y = np.meshgrid(rng,rng)
```

```python
def well(t,x):
    return [x[1],-g*x[1]+x[0]-x[0]**3.]
```

```python
def default_criterion(y):
    return True
```

```python
def time_check(t, t_end, dt):
    if dt < 0. and t_end < dt:
        return t+dt>t_end
    elif dt > 0. and t_end > dt:
        return t+dt<t_end
    else :
        print "Error"
    
```

```python
def integrate(x0, t_end=10, dt=0.1, criterion = default_criterion, option='endval'):
    
    if option == 'endval':
        r = ode(well).set_integrator('dopri5')
        r.set_initial_value(x0,0)
        while r.successful() and time_check(r.t,t_end,dt) and criterion(r.y):
            r.integrate(r.t+dt)
        return r.y
    elif option == 'orbit':
        x_final = []
        r = ode(well).set_integrator('dopri5')
        r.set_initial_value(x0,0)
        while r.successful() and time_check(r.t,t_end,dt) and criterion(r.y):
            r.integrate(r.t+dt)
            x_final.append(r.y)
        return x_final
```

```python
x_fin = []
Map = np.zeros_like(X)
for i in range(len(rng)):
    for j in range(len(rng)):
        x_fin.append(integrate([X[i,j],Y[i,j]], 10,0.1))
        Map[i,j] = x_fin[-1][0]
```

```python
X,G = np.meshgrid(rng,grng)
x_fin2 = []
Map2 = np.zeros_like(X)
for i in range(len(rng)):
    for j in range(len(rng)):
        g = G[i,j]
        x_fin2.append(integrate([X[i,j],0.], 10,0.1))
        Map2[i,j] = x_fin2[-1][0]
```

```python
fig2 = plt.figure()
ax2 = fig2.add_subplot(1,1,1)
ax2.set_ylabel('x0')
ax2.set_xlabel('g')
ax2.imshow(Map2.T, origin='lower', extent=[0,bound,-bound,bound])
```
