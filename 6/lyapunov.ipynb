```python
import sys
oldsysstdout = sys.stdout
class flushfile():
    def __init__(self, f):
        self.f = f
    def __getattr__(self,name): 
        return object.__getattribute__(self.f, name)
    def write(self, x):
        self.f.write(x)
        self.f.flush()
    def flush(self):
        self.f.flush()
sys.stdout = flushfile(sys.stdout)
```

```python
class defaultError(Exception):
    def __init__(self, value):
        self.value = value
    def __str__(self):
        return repr(self.value)
```

```python
#default methods now raise the defaultError
def default_criterion(t,y):
    raise defaultError("Error: No termination criterion provided!")
    return False
def default_function(x):
    raise defaultError("Error: No function provided!")
    return x
```

```python
def integrate(x0, t0=0, dt=0.1, criterion= default_criterion, function = default_function):
    orbit = [np.array([0,x0])]
    current_state = ode(function).set_integrator('dopri5')
    current_state.set_initial_value(x0,t0)
    while current_state.successful() and criterion(current_state.t,current_state.y):
        current_state.integrate(current_state.t+dt)
        orbit.append(np.array([current_state.t, current_state.y]))
    return orbit
```

```python
#Global Params
r = 3.5
dx = 0.0001
b = 3.
```

```python
def logistische_abbildung(x):
    return float(r)*x*(1.-x)
```

```python
def shift(x):
    return np.mod(b*x, 1.)
```

```python
def logistischer_orbit(x0,n_max):
    i = 0
    x = [x0]
    while i < n_max:
        x.append(logistische_abbildung(x[i]))
        i = i+1
    return np.array(x)
```

```python
x = logistischer_orbit(0.00001*np.random.random(2),10000)
lyapunov_exponent = 1./np.arange(1,len(x), 1.) * np.log(np.absolute(x[1:,0]-x[1:,1])/(np.absolute(x[0,0]-x[0,1])))
```

```python
fig = plt.figure() 
ax = fig.add_subplot(111) 
ax.set_xlabel('n') 
ax.set_ylabel('$\lambda(n)$') 
ax.set_title('Time evolution of the Lyapunov exponent') 
ax.scatter(np.arange(1,len(x)), lyapunov_exponent, color = 'black', s=0.1)
```

```python
lya_of_r = []
for r in np.linspace(2,4,100):
    lya = []
    for i in range(1000):
        rnd = np.random.random(1)
        rnd0 = np.array([rnd, np.mod(rnd+dx, 1.)])
        x = logistischer_orbit(rnd0,25)
        lya.append(1./float(len(x)) * np.log(np.absolute(x[-1,0]-x[-1,1])/np.absolute(x[0,0]-x[0,1])) )
    lya = np.array(lya)
    lya_of_r.append([r,
                     np.mean(lya[lya>-1e5])
                     ])
r=2.0
lya_of_r = np.array(lya_of_r)
```
