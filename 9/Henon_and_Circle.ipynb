```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline  
import matplotlib
from numpy.linalg import qr
import pylab
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
a = 1.4
b = 0.3
```

```python
def henon(x):
    return [1-a*x[0]*x[0]+b*x[1], x[0]]
```

```python
def null_jac(x):
    return np.matrix(
    [[-2*a*x[0], b],
     [   1     , 0]]
    )
```

```python
def flow_henon(t, x0):
    x = [x0]
    for i in range(t):
        x.append(henon(x[-1]))
    return np.array(x)
```

```python
N = 10000
```

```python
phi = flow_henon(N, [0.,0.])
```

```python
jacobian = []
for x in phi:
    jacobian.append(null_jac(x))
```

```python
Q,R = [],[]
Rii = []
for mat in jacobian:
    q,r = qr(mat)
    d = np.matrix(np.diag([np.sign(r[0,0]),np.sign(r[1,1])]))
    r = d*r
    Rii.append([r[0,0],r[1,1]])
```

```python
lyapunov = float(1./N)*np.sum(np.log(Rii),axis = 0)
print lyapunov
```

```python
def circle_map(theta, omega):
    return theta+omega-0.9/2/np.pi*np.sin(2*np.pi*theta)
circle_map = np.vectorize(circle_map)
```

```python
theta = np.linspace(0,1,100)
omega = np.linspace(0,1,100)
N = 1000
T,O = np.meshgrid(theta,omega)
T0 = T
for i in range(N):
    T = circle_map(T, O)
```

```python
winding_no = np.mean((T-T0)/N, axis=1)
```

```python
plt.scatter(O, np.mod(T, 1), s = 0.5)
plt.xlabel('$\Omega$')
plt.ylabel('$\theta$')
```

```python
plt.scatter(np.mean(O, axis = 1), winding_no, s = 1)
plt.xlabel('$\Omega$')
plt.ylabel('$W$')
```

```python
def dot_circle(x):
    return np.log(np.absolute(1.-0.9*np.cos(2*np.pi*x)))
dot_circle = np.vectorize(dot_circle)
```

```python
T = T0
SUM = np.zeros_like(T)
for i in range(1000):
    T = circle_map(T,O)
    SUM = SUM + dot_circle(T)
lyapunov = 1./1000. * SUM
print lyapunov
```
